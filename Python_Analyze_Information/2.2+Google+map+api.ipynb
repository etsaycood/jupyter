```python
areas = ["中正","大同","中山","松山","大安","萬華","信義","士林","北投","內湖","南港","文山"]
```

```python
ids = []
for areas in areas:
    # Geocoding an address
    geocode_result = gmaps.geocode(areas)
    loc = geocode_result[0]['geometry']['location']
    print(areas+"的夜市數量: "+str(len(gmaps.places_radar(keyword="夜市", location=loc, radius=25000)['results'])))
    for place in gmaps.places_radar(keyword="夜市", location=loc, radius=25000)['results']:
        ids.append(place['place_id'])
```

```python
stores_info = []
for id in ids:
    stores_info.append(gmaps.place(place_id=id, language='zh-TW')['result'])
```

```python
output = pd.DataFrame.from_dict(stores_info)
```
